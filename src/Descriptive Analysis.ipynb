```python
df = pd.read_csv(os.path.join(input_dir,r"welfare_data.csv"))
```

```python
# using EOY 2016 rating as quality measure

df_2017_nov = df[(df["year"]==2016)&(df["month"]==12)]
df_2017_nov["rating_2017_nov"] = df_2017_nov["Rating"]

df_year_hotel = df.groupby(["year","shareid"])["Demand"].sum().reset_index()


df_temp = df_year_hotel.merge(df_2017_nov[["shareid","rating_2017_nov"]], on="shareid", how="left")
df_temp = df_temp.dropna(subset=["rating_2017_nov"])
df_temp = df_temp[df_temp["rating_2017_nov"]!=0]
df_year = df_temp.groupby(["year"])["Demand"].sum().reset_index().rename(columns={"Demand":"demand_year_total"})
df_temp = df_temp.merge(df_year, on=["year"])
df_temp["weight_rating"] = (df_temp["Demand"]/df_temp["demand_year_total"])*df_temp["rating_2017_nov"]
df_temp.groupby("year")["weight_rating"].sum()
```
